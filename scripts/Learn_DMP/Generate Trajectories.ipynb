```python
acc = npy.zeros((100,2))
for t in range(100):
#     acc[t,0] = math.cos(float(t)/31.42)/900
#     acc[t,1] = math.sin(float(t)/31.42)/900
    acc[t,0] = math.cos(float(t)/31.42)/900
    acc[t,1] = math.sin(float(t)/31.42)/900
    
vel = npy.zeros((100,2))
pos = npy.zeros((100,2))
dt=1
for t in range(0,100):
    vel[t] = vel[t-1]+acc[t]*dt
    
for t in range(1,100):
    pos[t] = pos[t-1]+vel[t]*dt+acc[t]*(dt**2)
    
plt.plot(pos[:,0],pos[:,1])
plt.show()
```

```python
with file('acceleration.npy','w') as outfile:
    npy.save(outfile,acc)
    
with file('velocity.npy','w') as outfile:
    npy.save(outfile,vel)
    
with file('position.npy','w') as outfile:
    npy.save(outfile,pos)
    
```
